Iris dataset contains:
1. 3 clases
  *   setosa
  *   versicolor
  *   virginica
2. 150 total samples
3. 50 samples for each class
4. 4 features
  *   SepalLengthCm
  *   SepalWidthCm
  *   PetalLengthCm
  *   PetalWidthCm

```python
#Importing libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
import seaborn as sns
iris = load_iris()
```

```python
#Plot the histogram for each class

colors = ['blue', 'red', 'green']

for i in range(len(iris.feature_names)):
  plt.figure(figsize=(5,5))
  for label, color in zip(range(len(iris.target_names)),colors):
    plt.hist(iris.data[iris.target == label , i],
    label = iris.target_names[label], color = color)
  plt.legend(loc='upper right')
  plt.show()
```

```python
#Use scatter plot to plot every 2 attributes together

colors = ['blue', 'red']

for i in range(len(iris.feature_names)):
  for j in range(i+1, len(iris.feature_names)):
    for label, color in zip(range(len(iris.target_names)), colors):
      plt.scatter(iris.data[iris.target==label, i],
                  iris.data[iris.target==label, j],
                  label=iris.target_names[label],
                  alpha = 1,
                  color=color)
      plt.xlabel(iris.feature_names[i])
      plt.ylabel(iris.feature_names[j])
      plt.legend()
    plt.show()
```

```python
#Use 3D scatter plot to plot every 3 attributes together

colors = ['blue', 'red', 'green']
for i in range(len(iris.feature_names)):
  for j in range(i+1, len(iris.feature_names)):
    for k in range(j+1, len(iris.feature_names)):
      plt.figure(figsize = (16, 9))
      ax = plt.axes(projection ="3d")
      for label, color in zip(range(len(iris.target_names)),colors):
        plt.scatter(iris.data[iris.target==label, i],
        iris.data[iris.target==label, j],
        iris.data[iris.target==label, k],
        label=iris.target_names[label],
        alpha = 1,
        color=color)
        ax.set_xlabel(iris.feature_names[i])
        ax.set_ylabel(iris.feature_names[j])
        ax.set_zlabel(iris.feature_names[k])
        plt.legend()
      plt.show()
```

```python
#Show the correlation matrix between the attributes and comment on it
 
dataFrame = pd.DataFrame(iris.data, columns = iris.feature_names)
sns.heatmap(dataFrame.corr(), annot=True)
plt.show()
```

```python
#Use boxplots to show the quartiles and the median of the sepal length, sepal width, petal length and petal width for each kind of flowers.
 
for i in range(len(iris.feature_names)):
  sns.boxplot(x= iris.target,y=iris.data.T[i])
  plt.xticks([0,1,2],iris.target_names)
  plt.ylabel(iris.feature_names[i])
  plt.show()
```
