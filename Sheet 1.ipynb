```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.metrics.pairwise import euclidean_distances
```

```python
data = np.array([[10,60,10,90],[20,50,40,70],[30,50,30,40],[20,50,20,60],[10,60,30,10]])
print("Data set:")
print(data)
print("Norm:")
print(np.linalg.norm(data, axis=1))
print("Cosine similarity:")
print(cosine_similarity(data))
print("Euclidean distance:")
print(euclidean_distances(data))
```

```python
#Question 3

data = np.array([[10,60,90],[20,50,70],[30,50,40],[20,50,60],[10,60,10]])
feature_names = np.array(["a1","a2","a4"])
print("Data set:")
print(data)

fig = plt.figure(figsize = (10, 7))
ax = plt.axes(projection ="3d")
ax.scatter3D(data[:,0],data[:,1],data[:,2], color = "blue")
plt.title("3D scatter plot for D3")
print("Data set D3:")
print(data)
mean = np.mean(data,axis=0)
print("Mean:")
print(mean)
print("Centered data matrix:")
z = data - mean
print(z)
print("Covariance matrix:")
cov = np.cov(np.transpose(z),bias=True)
print(cov)
eValues,eVectors = np.linalg.eigh(cov)
eValues = np.flip(eValues)
eVectors = np.flip(eVectors,axis = 1)
print("Eigen values:")
print(eValues)
print("Eigen vectors:")
print(eVectors)
print("Verify UT˄U= COV.:")
print(np.dot(np.dot(eVectors,np.diag(eValues)),np.transpose(eVectors)))
print("Explained variance:")
print(eValues[0]/np.sum(eValues)*100)
p = np.delete(eVectors,2,1)
print("projection matrix")
print(p)
print("Project the instances into a 2-Dimension space")
d2 = np.transpose(np.dot(np.transpose(p),np.transpose(z)))
print(d2)
```
